# How to Save image or numpy as image
* **How to convert Video into images**

```python
import numpy as np
import cv2
import os
```

```python
path = 'Data/Image/'
os.listdir(path)
```

* **Save Numpy Array**

```python
rand_array = np.random.randint(255,size = (300,600,3))
rand_array
cv2.imwrite('Data/rand_np_array.png',rand_array) #
img = cv2.imread('Data/rand_np_array.png')
cv2.imshow('rand_np_array.png',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

* **Save Image from Video**

```python
path = 'Data/Video/'
os.listdir(path)
```

```python
video_path = path + '/'+'aakash.mp4'
os.mkdir('Data/Video_to_image') # Create directory
cap = cv2.VideoCapture(video_path)
img_count = 1
while cap.isOpened():
    ret,frame = cap.read()
    if not ret:
        print("Unable to read frame")
        break
    is_img_write = cv2.imwrite(f"Data/Video_to_image/aakash{img_count}.jpg",frame)
    if is_img_write:
        print(f"image save to Video_to_image/aakash{img_count}.jpg")
    cv2.imshow('video',frame)
    cv2.waitKey(25) & 0xff == ord('q')
    img_count+= 1
cap.release()
cv2.destroyAllWindows()
    
```
